# 파이토치 선형 회귀 실습

```
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
```

### 데이터 생성

```
x_train = np.array([[3.3], [4.4], [5.5], [6.71], [6.93], [4.168],
                    [9.779], [6.182], [7.59], [2.167], [7.042],
                    [10.791], [5.313], [7.997], [3.1]], dtype=np.float32)

y_train = np.array([[1.7], [2.76], [2.09], [3.19], [1.694], [1.573],
                    [3.366], [2.596], [2.53], [1.221], [2.827],
                    [3.465], [1.65], [2.904], [1.3]], dtype=np.float32)
```

### 데이터 확인

```
x_train.shape
```

```
y_train.shape
```

```
plt.plot(x_train, y_train, 'ro', label='Original data')
plt.legend()
plt.show()
```

### 파라미터 설정

```
# 하이퍼파라미터
input_size = 1
output_size = 1
num_epochs = 60
learning_rate = 0.001
```

### Linear regression 모델 정의

torch.nn.Linear 클래스 사용

$y=x*w+b$

https://pytorch.org/docs/master/nn.html?highlight=linear#torch.nn.Linear

```
# Linear regression model
model = nn.Linear(input_size, output_size)
```

### 손실함수와 옵티마이저

[MSELoss](https://pytorch.org/docs/master/generated/torch.nn.MSELoss.html#torch.nn.MSELoss)


[SGD](https://pytorch.org/docs/master/optim.html?highlight=optim%20sgd#torch.optim.SGD)

```
# Loss and optimizer
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

### 모델 학습(Model training)

```
# Train the model
for epoch in range(num_epochs): # 설정한 epoch 만큼 학습합니다.
    # numpy 배열에서 torch.Tensor로 전환
    inputs = torch.from_numpy(x_train)
    targets = torch.from_numpy(y_train)

    # Forward pass
    outputs = model(inputs)
    loss = criterion(outputs, targets)

    # Backward and optimize
    optimizer.zero_grad()  # 옵티마이저 초기화(이전 학습의 영향 제거)
    loss.backward()
    optimizer.step()

    if (epoch+1) % 5 == 0: # 5 epoch 마다 학습 상황이 출력됩니다.
        print ('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1, num_epochs, loss.item()))
```

#### 예측(Prediction)

```
predicted = model(torch.from_numpy(x_train)).detach().numpy()
```
